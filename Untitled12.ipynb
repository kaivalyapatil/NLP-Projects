**DataDIct creation**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15, stratify=y)

x_train_list = X_train['Keywordbasetable'].astype('str').to_list()
x_test_list = X_test['Keywordbasetable'].astype('str').to_list()

y_train_list, y_test_list = y_train.to_list(),y_test.to_list()

train_data = {"text" :x_train_list,"label":y_train_list}
test_data = {"text" :x_test_list,"label":y_test_list}
data_dict = DatasetDict({
    "train": Dataset.from_dict(train_data),
    "test": Dataset.from_dict(test_data)
})
```

**Tokenization**

```python
tokenizer = AutoTokenizer.from_pretrained('distilbert-base-uncased')
```

```python
def preprocess_function(examples):
    return tokenizer(examples['text'],truncation= True)


tokenized_dataset = data_dict.map(preprocess_function, batched = True)
```

**Padding**

```python
data_collector = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
id2label = {0: "NEGATIVE", 1: "POSITIVE"}
label2id = {"NEGATIVE": 0, "POSITIVE": 1}
```

**Model creation**

```python
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased",num_labels=2, id2label=id2label, label2id=label2id)
```

```python
from datasets import load_metric

metric  = load_metric("accuracy")

def compute_metric(eval_pred):
  logits , labels = eval_pred
  predictions = np.argmax(logits, axis=-1)
  return metric.compute(predictions=predictions, references=labels)


training_args = TrainingArguments(
    output_dir="./results",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=5,
    weight_decay=0.01,
    evaluation_strategy="epoch"
)

trainer = Trainer(
    model = model,
    args = training_args,
    data_collator = data_collector,
    train_dataset = tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["test"],
    tokenizer= tokenizer,
    compute_metrics = compute_metric
)

```

```python
trainer.train()
```

```python
trainer.save_model()
```
